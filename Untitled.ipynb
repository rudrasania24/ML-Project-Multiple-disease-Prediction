```python
# preprocessing
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

heart    = pd.read_csv('heart.csv')
dengue   = pd.read_csv('dengue.csv')
diabetes = pd.read_csv('diabetes.csv')


dengue = dengue.drop('Name', axis=1)


heart    = heart.fillna(heart.mean())
dengue   = dengue.fillna(dengue.mean())
diabetes = diabetes.fillna(diabetes.mean())

print("Missing values after filling:")
print("Heart   :", heart.isnull().sum().sum())
print("Dengue  :", dengue.isnull().sum().sum())
print("Diabetes:", diabetes.isnull().sum().sum())


# X = input features
# y = target column (what we predict)

X_heart    = heart.drop('target', axis=1)
y_heart    = heart['target']

X_dengue   = dengue.drop('Dengue', axis=1)
y_dengue   = dengue['Dengue']

X_diabetes = diabetes.drop('Outcome', axis=1)
y_diabetes = diabetes['Outcome']

print("\nX and y separated ✓")


# 80% training, 20% testing

X_train_h,  X_test_h,  y_train_h,  y_test_h  = train_test_split(
    X_heart,    y_heart,    test_size=0.2, random_state=42)

X_train_d,  X_test_d,  y_train_d,  y_test_d  = train_test_split(
    X_dengue,   y_dengue,   test_size=0.2, random_state=42)

X_train_db, X_test_db, y_train_db, y_test_db = train_test_split(
    X_diabetes, y_diabetes, test_size=0.2, random_state=42)

print("\nTrain Test Split ✓")
print("Heart    — Train:", X_train_h.shape,  "Test:", X_test_h.shape)
print("Dengue   — Train:", X_train_d.shape,  "Test:", X_test_d.shape)
print("Diabetes — Train:", X_train_db.shape, "Test:", X_test_db.shape)


# Makes all numbers in same range

scaler_h  = StandardScaler()
scaler_d  = StandardScaler()
scaler_db = StandardScaler()

X_train_h  = scaler_h.fit_transform(X_train_h)
X_test_h   = scaler_h.transform(X_test_h)

X_train_d  = scaler_d.fit_transform(X_train_d)
X_test_d   = scaler_d.transform(X_test_d)

X_train_db = scaler_db.fit_transform(X_train_db)
X_test_db  = scaler_db.transform(X_test_db)

print("\nScaling Done ✓")
print("\nPreprocessing Complete ✓")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

heart    = pd.read_csv('heart.csv')
dengue   = pd.read_csv('dengue.csv')
diabetes = pd.read_csv('diabetes.csv')

dengue = dengue.drop('Name', axis=1)

heart    = heart.fillna(heart.mean())
dengue   = dengue.fillna(dengue.mean())
diabetes = diabetes.fillna(diabetes.mean())

X_heart    = heart.drop('target', axis=1)
y_heart    = heart['target']

X_dengue   = dengue.drop('Dengue', axis=1)
y_dengue   = dengue['Dengue']

X_diabetes = diabetes.drop('Outcome', axis=1)
y_diabetes = diabetes['Outcome']

heart_cols    = X_heart.columns.tolist()
dengue_cols   = X_dengue.columns.tolist()
diabetes_cols = X_diabetes.columns.tolist()

X_train_h,  X_test_h,  y_train_h,  y_test_h  = train_test_split(X_heart,    y_heart,    test_size=0.2, random_state=42)
X_train_d,  X_test_d,  y_train_d,  y_test_d  = train_test_split(X_dengue,   y_dengue,   test_size=0.2, random_state=42)
X_train_db, X_test_db, y_train_db, y_test_db = train_test_split(X_diabetes, y_diabetes, test_size=0.2, random_state=42)

scaler_h  = StandardScaler()
scaler_d  = StandardScaler()
scaler_db = StandardScaler()

X_train_h  = scaler_h.fit_transform(X_train_h)
X_test_h   = scaler_h.transform(X_test_h)

X_train_d  = scaler_d.fit_transform(X_train_d)
X_test_d   = scaler_d.transform(X_test_d)

X_train_db = scaler_db.fit_transform(X_train_db)
X_test_db  = scaler_db.transform(X_test_db)

print("Preprocessing Done ✓")

rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(X_train_h, y_train_h)
heart_importance = pd.Series(rf.feature_importances_, index=heart_cols).sort_values(ascending=False)
print("\n=== HEART — Feature Importance ===")
print(heart_importance)

rf.fit(X_train_d, y_train_d)
dengue_importance = pd.Series(rf.feature_importances_, index=dengue_cols).sort_values(ascending=False)
print("\n=== DENGUE — Feature Importance ===")
print(dengue_importance)

rf.fit(X_train_db, y_train_db)
diabetes_importance = pd.Series(rf.feature_importances_, index=diabetes_cols).sort_values(ascending=False)
print("\n=== DIABETES — Feature Importance ===")
print(diabetes_importance)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

heart_importance.plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='black')
axes[0].set_title('Heart — Feature Importance')
axes[0].set_ylabel('Score')
axes[0].tick_params(axis='x', rotation=45)

dengue_importance.plot(kind='bar', ax=axes[1], color='tomato', edgecolor='black')
axes[1].set_title('Dengue — Feature Importance')
axes[1].set_ylabel('Score')
axes[1].tick_params(axis='x', rotation=45)

diabetes_importance.plot(kind='bar', ax=axes[2], color='mediumseagreen', edgecolor='black')
axes[2].set_title('Diabetes — Feature Importance')
axes[2].set_ylabel('Score')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

heart_selected    = heart_importance[heart_importance > 0.05].index.tolist()
dengue_selected   = dengue_importance[dengue_importance > 0.05].index.tolist()
diabetes_selected = diabetes_importance[diabetes_importance > 0.05].index.tolist()

print("\n=== SELECTED FEATURES ===")
print("Heart    :", heart_selected)
print("Dengue   :", dengue_selected)
print("Diabetes :", diabetes_selected)

print("\nFeature Engineering Done ✓")
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

#  LOAD DATA
heart    = pd.read_csv('heart.csv')
dengue   = pd.read_csv('dengue.csv')
diabetes = pd.read_csv('diabetes.csv')

print("Heart shape   :", heart.shape)
print("Dengue shape  :", dengue.shape)
print("Diabetes shape:", diabetes.shape)

# PREPROCESSING 
dengue   = dengue.drop('Name', axis=1)
heart    = heart.fillna(heart.mean())
dengue   = dengue.fillna(dengue.mean())
diabetes = diabetes.fillna(diabetes.mean())

X_heart    = heart.drop('target', axis=1)
y_heart    = heart['target']
X_dengue   = dengue.drop('Dengue', axis=1)
y_dengue   = dengue['Dengue']
X_diabetes = diabetes.drop('Outcome', axis=1)
y_diabetes = diabetes['Outcome']

heart_cols    = X_heart.columns.tolist()
dengue_cols   = X_dengue.columns.tolist()
diabetes_cols = X_diabetes.columns.tolist()

X_train_h,  X_test_h,  y_train_h,  y_test_h  = train_test_split(X_heart,    y_heart,    test_size=0.2, random_state=42)
X_train_d,  X_test_d,  y_train_d,  y_test_d  = train_test_split(X_dengue,   y_dengue,   test_size=0.2, random_state=42)
X_train_db, X_test_db, y_train_db, y_test_db = train_test_split(X_diabetes, y_diabetes, test_size=0.2, random_state=42)

scaler_h  = StandardScaler()
scaler_d  = StandardScaler()
scaler_db = StandardScaler()

X_train_h  = scaler_h.fit_transform(X_train_h)
X_test_h   = scaler_h.transform(X_test_h)
X_train_d  = scaler_d.fit_transform(X_train_d)
X_test_d   = scaler_d.transform(X_test_d)
X_train_db = scaler_db.fit_transform(X_train_db)
X_test_db  = scaler_db.transform(X_test_db)

print("Preprocessing Done ✓")

#  MODEL TRAINING 
models = {
    "Logistic Regression" : LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest"       : RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting"   : GradientBoostingClassifier(n_estimators=100, random_state=42),
    "SVM"                 : SVC(probability=True, random_state=42)
}

print("\n HEART DISEASE ")
heart_trained = {}
for name, model in models.items():
    model.fit(X_train_h, y_train_h)
    heart_trained[name] = model
    print(f"{name} trained ✓")

print("\n DENGUE ")
dengue_trained = {}
for name, model in models.items():
    model.fit(X_train_d, y_train_d)
    dengue_trained[name] = model
    print(f"{name} trained ✓")

print("\n DIABETES")
diabetes_trained = {}
for name, model in models.items():
    model.fit(X_train_db, y_train_db)
    diabetes_trained[name] = model
    print(f"{name} trained ✓")

print("\nAll Models Trained ✓")

# QUICK TEST
print("\n QUICK TEST ")

sample_heart = X_test_h[0].reshape(1, -1)
pred = heart_trained["Random Forest"].predict(sample_heart)
prob = heart_trained["Random Forest"].predict_proba(sample_heart)[0][1]
print("\nHeart Disease")
print("  Prediction  :", "POSITIVE" if pred[0] == 1 else "NEGATIVE")
print("  Probability :", round(prob * 100, 2), "%")

sample_dengue = X_test_d[0].reshape(1, -1)
pred = dengue_trained["Random Forest"].predict(sample_dengue)
prob = dengue_trained["Random Forest"].predict_proba(sample_dengue)[0][1]
print("\nDengue")
print("  Prediction  :", "POSITIVE" if pred[0] == 1 else "NEGATIVE")
print("  Probability :", round(prob * 100, 2), "%")

sample_diabetes = X_test_db[0].reshape(1, -1)
pred = diabetes_trained["Random Forest"].predict(sample_diabetes)
prob = diabetes_trained["Random Forest"].predict_proba(sample_diabetes)[0][1]
print("\nDiabetes")
print("  Prediction  :", "POSITIVE" if pred[0] == 1 else "NEGATIVE")
print("  Probability :", round(prob * 100, 2), "%")

print("\nAll Done ✓")
```
